```python
def get_k_similar_users(userID, books_matrix, k):
  input_user_ratings = books_matrix[books_matrix['user_id'] == userID].set_index('isbn')  # input users rating


  other_users_ratings = books_matrix[books_matrix['user_id'] != userID]  # other users rating


  common_books = input_user_ratings.index.intersection(other_users_ratings['isbn'].unique())  # common isbn


  if not common_books.empty:

    input_user_ratings = input_user_ratings[input_user_ratings.index.isin(common_books)] #common isbn user id with rating

    other_users_ratings = other_users_ratings[other_users_ratings['isbn'].isin(common_books)]

    ratings_matrix = other_users_ratings.pivot_table(index='user_id', columns='isbn', values='rating', fill_value=0)  # matrix for common ratring

    similarity = cosine_similarity([input_user_ratings['rating'].tolist()], ratings_matrix)  # cosin distance

    #print(f'similarity\n {similarity}')

    similar_users_similarity = pd.DataFrame({'user_id': ratings_matrix.index, 'similarity': similarity[0]}) #df with usier id and similarity, at 0th has inputID s similarity

    similar_users_similarity = similar_users_similarity.sort_values(by='similarity', ascending=False)

    k_users = similar_users_similarity.head(k)
    return k_users
  else:
      print(f"Common book not found for user {userID}")
      return pd.DataFrame()
```

```python
def predict_book_rating(userID, books_matrix, similar_users, isbn):

  input_user_ratings = books_matrix[(books_matrix['user_id'] == userID) & (books_matrix['isbn'] == isbn)]['rating'].values

  if len(input_user_ratings)>0:
    return f"Already rated and rating {input_user_ratings[0]}"

  similar_user_ratings = books_matrix[(books_matrix['user_id'].isin(similar_users['user_id'])) & (books_matrix['isbn'] == isbn)]

  user_similarity = pd.merge(similar_user_ratings, similar_users, on='user_id') #user id, isbn, rating, similarity

  if similar_user_ratings.empty:
    return "Similar ratings not found!"

  total_similarity = user_similarity['similarity'].sum()

  total = 0
  for i in range(0, len(user_similarity)):
    total = total + (user_similarity['rating'][i]*user_similarity['similarity'][i])

  rating = total/total_similarity

  return rating
```

```python
def recommendation(userID, books_info, books_matrix):
  a = books_matrix[books_matrix['user_id']==userID]
  if len(a)==0:
    return f'User ID not found!'

  top_similar_users = get_k_similar_users(userID, books_matrix, 15)
  usr_id = top_similar_users['user_id']
  similar_user_set = books_info[books_info['user_id'].isin(usr_id)]
  unique_books = similar_user_set['isbn'].unique()
  total_reating = similar_user_set[similar_user_set['isbn'].isin(unique_books)]
  total_reating.drop_duplicates(['isbn'], inplace = True)
  total_reating = total_reating.sort_values(by = 'total_rating', ascending = False)
  r = total_reating[total_reating['user_id'] != userID]
  recomendation = r.head(10)
  ans = recomendation[['isbn','book_title','book_author']]
  return ans
```
